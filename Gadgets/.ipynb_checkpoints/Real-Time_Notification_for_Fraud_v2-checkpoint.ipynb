```python
#import libararies
import io 
import pandas as pd
from tabulate import tabulate
from sqlalchemy import create_engine
import datetime
#connect to mysql
import pymysql
#Send out emails:
import smtplib
from email import encoders
from email.mime.text import MIMEText
from email.mime.base import MIMEBase
from email.mime.multipart import MIMEMultipart

#define notification function
def notification():
    
    #define is_empty funtion to tell if a structure empty or not
    def is_empty(any_structure):
        if any_structure:
            return False
        else:
            return True
        
    #Query to detect new status change
    query="""""""""
            select distinct
                a.* 
            from 
            (
                your query to detect status change
            ) a
            left join Status_Change_Hist b 
            on a.ID=b.ID
            where b.ID is null
          """""""""
    
    #Establish a Mysql connection:
    with pymysql.connect(host='yourhost',user='yourusername',password='yourpwd',db='yourdb',port=3306) as con:
        #Detect status change by running query
        con.execute(query)
        result = con.fetchall()
        data = []
        for row in result:
            data.append(row)
        if is_empty(result)==False: #Status change detected if the result is not empty:
            
            #Convert data to DataFrame
            df = pd.DataFrame(data, columns =['-','-']) #Rename columns if you want
            df.index=df.index+1 #Start row index from 1 instead of 0
            
            #Send out notification via email if Status change detected
            #Body text
            txt = """
                Hello Team,

                The records below were detected as status change:

                {table}

                Regards,

                Andy Chen
                """
            html = """
                <html>
                <head>
                <style> 
                 table, th, td {{ border: 1px solid black; border-collapse: collapse; }}
                  th, td {{ padding: 5px; }}
                </style>
                </head>
                <body><p>Hello Team,</p>
                <p>The records below were detected as status change:</p>
                {table}
                <p>Regards,</p>
                <p>Andy Chen</p>
                </body></html>
                """
            txt = txt.format(table=tabulate(df, headers=df.columns, tablefmt="grid"))
            html = html.format(table=tabulate(df, headers=df.columns, tablefmt="html"))
            msg = MIMEMultipart("alternative", None, [MIMEText(txt), MIMEText(html,'html')])
            subject="Status Change Detected "+str(datetime.datetime.now())
            SENDER = "andy@google.com"
            RECIPIENTS=['andychen@google.com']
            #Construct the email
            msg['Subject'] = subject
            msg['From'] = SENDER
            msg['To'] = ", ".join(RECIPIENTS)
            s = smtplib.SMTP('host', port=25)
            #Send the email 
            s.sendmail(SENDER,RECIPIENTS, msg.as_string())
            s.close()
            print('emails sent out '+str(datetime.datetime.now()))
            
            #Insert new status change into Status_Change_Hist
            sql_insert ="""""""""
                        insert into Status_Change_HIst (columns)
                        select distinct
                            a.* 
                        from 
                        (
                            your query to detect status change
                        ) a
                        left join Status_Change_Hist b 
                        on a.ID=b.ID
                        where b.ID is null
                        """""""""
            with pymysql.connect(host='yourhost',user='yourusername',password='yourpwd',db='yourdb',port=3306) as con:
                con.execute(sql_insert)
            print('insert into Status_Change_Hist successfully '+str(datetime.datetime.now()))
        else:
            print('no emails sent out '+str(datetime.datetime.now()))

# Schedule Library imported 
import schedule 
import time 
schedule.every(1).seconds.do(notification) 
while True:
    schedule.run_pending()
    time.sleep(1)
```
